```python
fig = plt.figure()
ax = fig.gca(projection='3d')

# Create function to plot
X, Y = np.mgrid[-2:2:30j, -2:2:30j]
# Z = np.sin(np.pi*X)*np.sin(np.pi*Y)
Z = 7*X*Y/(np.exp(X**2 + Y**2))

# Create plot with wireframe on top
ax.plot_surface(X, Y, Z, cmap= "viridis", linewidth=2, rstride=1, cstride=1, alpha = 1.)
# ax.plot_wireframe(X, Y, Z, color='black', linewidth = .2)

# project contours onto 'wall' or 'floor' -- not working yet
#ax.contour(X, Y, Z, 10, linestyles="solid")


ax.view_init(elev = 60, azim = 75)
fig.set_size_inches(9, 6)
fig.savefig("images/surfaceplt.png")
plt.show()
```
